# Fuerzas de fricción

## Objetivos

1. Verificar mediante una simulación en computadora la acción de las fuerzas de fricción estática
y dinámica para un objeto que se desliza en una superficie.

2. Verificar el coeficiente de fricción de dos superficies con dos diferentes métodos.


# Procedimiento

En esta práctica virtual se hará uso de la simulación que puede ejecutarse o descargarse de la
dirección:
https://phet.colorado.edu/es/simulation/legacy/ramp-forces-and-motion

##Pasos
1. Explore la simulación y familiarícese con su funcionamiento.

2. Qué cálculos son necesarios para medir el coeficiente de fricción entre dos superficies. Realice
el cálculo equivalente con la simulación.

3. Calcule el coeficiente de fricción estático entre la superficie de madera y el cajón pequeño:
a) Aplicando la fuerza máxima sin que se mueva en la parte plana. (Método 1)
b) Levantando al ángulo máximo sin que la caja se mueva. (Método 2)
c) Capture las pantallas que sean necesarias. Vea que puede grabar la acción y luego
repetirla para verificarla.

4. Calcule ahora el coeficiente de fricción cinética siguiendo los siguientes pasos:
a) “Mida” que fuerza es la que hace que el cajón se mueva con rapidez constante con la
rampa totalmente horizontal. (Método 1)
b) Determine el ángulo para el cual luego de un leve empujón la caja baja con rapidez
constante. (Método 2)
c) Capture también para su reporte las pantallas que sean necesarias.

## Nota
1. Explique cómo verificó el coeficiente de fricción estático y cuánto es, por el método 1 y por el
método 2. Realice los cálculos.

2. Explique cómo verificó el coeficiente de fricción cinético y cuánto es, por el método 1 y por el
método 2. Realice los cálculos.

3. Qué puede concluir respecto de los dos coeficientes? ¿y de las fuerzas de fricción?
